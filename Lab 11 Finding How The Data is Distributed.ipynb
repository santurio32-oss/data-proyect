### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# --- 1Ô∏è‚É£ Ver estructura general del dataset ---
print("üîπ Shape del dataset:", df.shape)
print("\nüîπ Primeras filas del dataset:")
display(df.head())

# --- 2Ô∏è‚É£ Ver columnas y tipos de datos ---
print("\nüîπ Columnas y tipos de datos:")
print(df.dtypes)

# --- 3Ô∏è‚É£ Informaci√≥n general (valores nulos, memoria, etc.) ---
print("\nüîπ Informaci√≥n del dataset:")
print(df.info())

# --- 4Ô∏è‚É£ Resumen estad√≠stico de columnas num√©ricas ---
print("\nüîπ Resumen estad√≠stico:")
display(df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
print("üîπ Valores faltantes por columna:")
missing_values = df.isnull().sum()
display(missing_values[missing_values > 0])

# --- 2Ô∏è‚É£ Ver porcentaje de valores faltantes ---
print("\nüîπ Porcentaje de valores faltantes por columna:")
missing_percent = (df.isnull().mean() * 100).round(2)
display(missing_percent[missing_percent > 0])
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mode()[0], inplace=True)
        print("\n‚úÖ Valores faltantes despu√©s del tratamiento:")
display(df.isnull().sum().sum())
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# --- 1Ô∏è‚É£ Verificar si las columnas existen ---
cols_to_check = ['Employment', 'JobSat', 'YearsCodePro']
for col in cols_to_check:
    if col not in df.columns:
        print(f"‚ö†Ô∏è La columna '{col}' no existe en el DataFrame.")
    else:
        # --- 2Ô∏è‚É£ Mostrar conteo de valores ---
        print(f"\nüîπ Distribuci√≥n de valores en '{col}':")
        print(df[col].value_counts(dropna=False))
        print("-" * 50)
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1Ô∏è‚É£ Verificar si existe JobSat o columnas relacionadas ---
if 'JobSat' in df.columns:
    # Usar directamente la columna JobSat
    job_sat_data = df['JobSat'].dropna()
    title = 'Distribuci√≥n de satisfacci√≥n laboral (JobSat)'
else:
    # Buscar columnas con JobSatPoints_
    job_sat_cols = [col for col in df.columns if 'JobSatPoints_' in col]

    if len(job_sat_cols) == 0:
        raise KeyError("‚ö†Ô∏è No se encontr√≥ ninguna columna relacionada con JobSat o JobSatPoints.")
    
    # Calcular promedio general de satisfacci√≥n
    df['JobSat_Overall'] = df[job_sat_cols].mean(axis=1)
    job_sat_data = df['JobSat_Overall'].dropna()
    title = 'Distribuci√≥n promedio de satisfacci√≥n laboral (JobSat_Overall)'

# --- 2Ô∏è‚É£ Crear KDE Plot ---
plt.figure(figsize=(8,5))
sns.kdeplot(job_sat_data, fill=True, color='cornflowerblue')
plt.title(title)
plt.xlabel('Nivel de satisfacci√≥n laboral')
plt.ylabel('Densidad')
plt.show()

# --- 3Ô∏è‚É£ Crear Pie Chart (agrupando en niveles) ---
if job_sat_data.dtype != 'O':  # Si es num√©rica
    # Crear rangos de satisfacci√≥n
    bins = [0, 2, 4, 6, 8, 10]
    labels = ['Muy baja', 'Baja', 'Media', 'Alta', 'Muy alta']
    categories = pd.cut(job_sat_data, bins=bins, labels=labels, include_lowest=True)
else:
    categories = job_sat_data  # Si es categ√≥rica (texto)

# Pie chart
plt.figure(figsize=(6,6))
categories.value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Blues", len(categories.unique())))
plt.title('Distribuci√≥n porcentual de satisfacci√≥n laboral')
plt.ylabel('')
plt.show()

# --- 4Ô∏è‚É£ Interpretaci√≥n autom√°tica ---
if job_sat_data.dtype != 'O':
    mean_score = job_sat_data.mean()
    print(f"üìà Promedio general de satisfacci√≥n laboral: {mean_score:.2f}")

    if mean_score >= 8:
        print("üí° La mayor√≠a de los profesionales muestran una **alta satisfacci√≥n laboral**.")
    elif mean_score >= 6:
        print("üôÇ La satisfacci√≥n laboral es **moderadamente alta**, aunque hay margen de mejora.")
    elif mean_score >= 4:
        print("üòê La satisfacci√≥n laboral es **media o mixta**, con opiniones divididas.")
    else:
        print("‚ö†Ô∏è Los niveles de satisfacci√≥n laboral son **bajos**, se√±alando posibles √°reas problem√°ticas.")
else:
    print("üóÇÔ∏è Distribuci√≥n categ√≥rica mostrada ‚Äî la mayor√≠a de respuestas se concentran en las categor√≠as m√°s frecuentes.")
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
from matplotlib_venn import venn2  # pip install matplotlib-venn

# --- 1Ô∏è‚É£ Asegurarse de que las columnas existen ---
cols_needed = ['LanguageHaveWorkedWith', 'LanguageWantToWorkWith']
for col in cols_needed:
    if col not in df.columns:
        raise KeyError(f"‚ö†Ô∏è La columna '{col}' no se encuentra en el DataFrame.")

# --- 2Ô∏è‚É£ Separar los lenguajes por fila ---
def split_languages(series):
    """Convierte los valores separados por ';' en listas y aplana."""
    all_langs = series.dropna().apply(lambda x: x.split(';'))
    flat_list = [lang.strip() for sublist in all_langs for lang in sublist]
    return flat_list

have_worked = split_languages(df['LanguageHaveWorkedWith'])
want_to_work = split_languages(df['LanguageWantToWorkWith'])

# --- 3Ô∏è‚É£ Contar frecuencia ---
have_counter = Counter(have_worked)
want_counter = Counter(want_to_work)

# --- 4Ô∏è‚É£ Crear DataFrame para gr√°fico de barras ---
all_languages = list(set(list(have_counter.keys()) + list(want_counter.keys())))
freq_df = pd.DataFrame({
    'Language': all_languages,
    'HaveWorkedWith': [have_counter.get(lang, 0) for lang in all_languages],
    'WantToWorkWith': [want_counter.get(lang, 0) for lang in all_languages]
})

# Ordenar por frecuencia en "HaveWorkedWith"
freq_df = freq_df.sort_values('HaveWorkedWith', ascending=False)

# --- 5Ô∏è‚É£ Gr√°fico de barras agrupadas ---
import numpy as np

x = np.arange(len(freq_df))
width = 0.4

plt.figure(figsize=(12,6))
plt.bar(x - width/2, freq_df['HaveWorkedWith'], width=width, label='Have Worked With', color='skyblue')
plt.bar(x + width/2, freq_df['WantToWorkWith'], width=width, label='Want To Work With', color='salmon')
plt.xticks(x, freq_df['Language'], rotation=45, ha='right')
plt.ylabel('N√∫mero de respuestas')
plt.title('Comparaci√≥n de lenguajes de programaci√≥n: usados vs deseados')
plt.legend()
plt.tight_layout()
plt.show()

# --- 6Ô∏è‚É£ Opcional: Diagrama de Venn para top 10 lenguajes ---
top_have = set([lang for lang, _ in have_counter.most_common(10)])
top_want = set([lang for lang, _ in want_counter.most_common(10)])

plt.figure(figsize=(6,6))
venn2([top_have, top_want], set_labels=('Have Worked With', 'Want To Work With'))
plt.title('Diagrama de Venn - Top 10 lenguajes')
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1Ô∏è‚É£ Verificar que las columnas necesarias existan ---
required_cols = ['RemoteWork', 'Country']  # Ajusta 'Country' si tu columna de regi√≥n se llama distinto
for col in required_cols:
    if col not in df.columns:
        raise KeyError(f"‚ö†Ô∏è La columna '{col}' no se encuentra en el DataFrame.")

# --- 2Ô∏è‚É£ Contar distribuci√≥n de RemoteWork por pa√≠s/regi√≥n ---
remote_counts = df.groupby(['Country', 'RemoteWork']).size().unstack(fill_value=0)
remote_counts = remote_counts.sort_values(remote_counts.columns[0], ascending=False)  # Ordenar por la primera categor√≠a

# --- 3Ô∏è‚É£ Gr√°fico de barras agrupadas ---
remote_counts.plot(kind='bar', figsize=(14,6), stacked=False)
plt.title('Distribuci√≥n de trabajo remoto por pa√≠s')
plt.ylabel('N√∫mero de encuestados')
plt.xlabel('Pa√≠s')
plt.xticks(rotation=45, ha='right')
plt.legend(title='RemoteWork')
plt.tight_layout()
plt.show()

# --- 4Ô∏è‚É£ Heatmap (opcional) ---
plt.figure(figsize=(12,6))
sns.heatmap(remote_counts, annot=True, fmt='d', cmap='Blues')
plt.title('Heatmap de trabajo remoto por pa√≠s')
plt.ylabel('Pa√≠s')
plt.xlabel('Tipo de RemoteWork')
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr

# --- 1Ô∏è‚É£ Buscar columnas relacionadas con satisfacci√≥n laboral ---
job_sat_cols = [col for col in df.columns if 'sat' in col.lower()]
if len(job_sat_cols) == 0:
    raise KeyError("‚ö†Ô∏è No se encontraron columnas relacionadas con satisfacci√≥n laboral.")

# --- 2Ô∏è‚É£ Crear columna promedio si hay varias columnas num√©ricas ---
numeric_cols = [col for col in job_sat_cols if pd.api.types.is_numeric_dtype(df[col])]
if len(numeric_cols) > 1:
    df['JobSat_Overall'] = df[numeric_cols].mean(axis=1)
    job_sat_series = df['JobSat_Overall']
else:
    job_sat_series = df[numeric_cols[0]] if numeric_cols else df[job_sat_cols[0]]

# --- 3Ô∏è‚É£ Asegurarse de que YearsCodePro exista y limpiar datos ---
if 'YearsCodePro' not in df.columns:
    raise KeyError("‚ö†Ô∏è La columna 'YearsCodePro' no se encuentra en el DataFrame.")
experience_series = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# --- 4Ô∏è‚É£ Eliminar filas con valores nulos ---
data = pd.concat([job_sat_series, experience_series], axis=1).dropna()
data.columns = ['JobSat', 'YearsCodePro']

# --- 5Ô∏è‚É£ Calcular correlaci√≥n ---
pearson_corr, pearson_p = pearsonr(data['JobSat'], data['YearsCodePro'])
spearman_corr, spearman_p = spearmanr(data['JobSat'], data['YearsCodePro'])

print(f"üìä Correlaci√≥n de Pearson: {pearson_corr:.3f} (p-value={pearson_p:.3f})")
print(f"üìä Correlaci√≥n de Spearman: {spearman_corr:.3f} (p-value={spearman_p:.3f})")

# --- 6Ô∏è‚É£ Visualizaci√≥n con scatter plot y l√≠nea de regresi√≥n ---
plt.figure(figsize=(8,6))
sns.scatterplot(data=data, x='YearsCodePro', y='JobSat', alpha=0.5)
sns.regplot(data=data, x='YearsCodePro', y='JobSat', scatter=False, color='red')
plt.title('Relaci√≥n entre experiencia profesional y satisfacci√≥n laboral')
plt.xlabel('A√±os de experiencia profesional (YearsCodePro)')
plt.ylabel('Satisfacci√≥n laboral promedio (JobSat)')
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# --- 1Ô∏è‚É£ Verificar que las columnas existan ---
required_cols = ['Employment', 'EdLevel']
for col in required_cols:
    if col not in df.columns:
        raise KeyError(f"‚ö†Ô∏è La columna '{col}' no se encuentra en el DataFrame.")

# --- 2Ô∏è‚É£ Crear tabla de contingencia ---
employment_vs_edu = pd.crosstab(df['EdLevel'], df['Employment'])

print("üìä Cross-tabulation: Education Level vs Employment")
print(employment_vs_edu)

# --- 3Ô∏è‚É£ Gr√°fico de barras apiladas ---
employment_vs_edu.plot(
    kind='bar',
    stacked=True,
    figsize=(12,6),
    colormap='tab20'
)
plt.title('Distribuci√≥n de tipo de empleo por nivel educativo')
plt.xlabel('Nivel educativo (EdLevel)')
plt.ylabel('N√∫mero de encuestados')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment')
plt.tight_layout()
plt.show()
```
