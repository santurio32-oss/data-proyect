### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
## Write your code here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# --- Conexión a la base de datos ---
conn = sqlite3.connect("survey-data.sqlite")

# --- Leer columnas de interés ---
query = """
SELECT Age, ConvertedCompYearly
FROM main
WHERE Age IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# --- Convertir edades a valores numéricos ---
def convert_age(age_str):
    try:
        if 'Under' in age_str:
            return 18
        elif '-' in age_str:
            parts = age_str.split('-')
            start = int(parts[0])
            end = int(parts[1].split(' ')[0])
            return (start + end) / 2
        else:
            return int(age_str.split(' ')[0])
    except:
        return None

df['AgeNum'] = df['Age'].apply(convert_age)

# --- Eliminar valores nulos ---
df = df.dropna(subset=['AgeNum', 'ConvertedCompYearly'])

# --- Crear grupos de edad ---
bins = [0, 24, 34, 44, 54, 64, 100]
labels = ['<25', '25-34', '35-44', '45-54', '55-64', '65+']
df['AgeGroup'] = pd.cut(df['AgeNum'], bins=bins, labels=labels)

# --- Calcular mediana de salario por grupo de edad ---
median_comp = df.groupby('AgeGroup')['ConvertedCompYearly'].median()

# --- Graficar línea ---
plt.figure(figsize=(10,6))
median_comp.plot(marker='o', linestyle='-', color='b')
plt.title("Median ConvertedCompYearly by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Median Annual Compensation (USD)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# --- Cerrar conexión ---
conn.close()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
## Write your code here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# --- Conexión a la base de datos ---
conn = sqlite3.connect("survey-data.sqlite")

# --- Leer columnas de interés ---
query = """
SELECT Age, ConvertedCompYearly
FROM main
WHERE Age IS NOT NULL AND ConvertedCompYearly IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# --- Convertir edades a valores numéricos ---
def convert_age(age_str):
    try:
        if 'Under' in age_str:
            return 18
        elif '-' in age_str:
            parts = age_str.split('-')
            start = int(parts[0])
            end = int(parts[1].split(' ')[0])
            return (start + end) / 2
        else:
            return int(age_str.split(' ')[0])
    except:
        return None

df['AgeNum'] = df['Age'].apply(convert_age)

# --- Filtrar edades entre 25 y 45 ---
df_filtered = df[(df['AgeNum'] >= 25) & (df['AgeNum'] <= 45)]
df_filtered = df_filtered.dropna(subset=['ConvertedCompYearly'])

# --- Crear grupos de edad ---
bins = [25, 34, 44, 45]  # rangos para este enfoque
labels = ['25-34', '35-44', '45']
df_filtered['AgeGroup'] = pd.cut(df_filtered['AgeNum'], bins=bins, labels=labels, right=True, include_lowest=True)

# --- Calcular mediana de salario por grupo de edad ---
median_comp_filtered = df_filtered.groupby('AgeGroup')['ConvertedCompYearly'].median()

# --- Graficar línea ---
plt.figure(figsize=(8,5))
median_comp_filtered.plot(marker='o', linestyle='-', color='green')
plt.title("Median ConvertedCompYearly for Ages 25 to 45")
plt.xlabel("Age Group")
plt.ylabel("Median Annual Compensation (USD)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# --- Cerrar conexión ---
conn.close()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
## Write your code here
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt

# --- Conexión a la base de datos ---
conn = sqlite3.connect("survey-data.sqlite")

# --- Leer columnas de interés ---
query = """
SELECT YearsCodePro, JobSatPoints_6
FROM main
WHERE YearsCodePro IS NOT NULL AND JobSatPoints_6 IS NOT NULL
"""
df = pd.read_sql_query(query, conn)

# --- Convertir YearsCodePro a numérico ---
def convert_years(x):
    try:
        if 'Less than' in x:
            return 0.5
        elif 'More than' in x:
            return int(x.split(' ')[2])  # "More than 50 years" -> 50
        else:
            return float(x)
    except:
        return None

df['YearsNum'] = df['YearsCodePro'].apply(convert_years)
df = df.dropna(subset=['YearsNum', 'JobSatPoints_6'])

# --- Crear grupos de experiencia ---
bins = [0, 2, 5, 10, 20, 50, 60]
labels = ['0-2','3-5','6-10','11-20','21-50','50+']
df['ExpGroup'] = pd.cut(df['YearsNum'], bins=bins, labels=labels, include_lowest=True)

# --- Calcular mediana de JobSatPoints_6 por grupo ---
median_jobsat = df.groupby('ExpGroup')['JobSatPoints_6'].median()

# --- Graficar línea ---
plt.figure(figsize=(10,6))
median_jobsat.plot(marker='o', linestyle='-', color='orange')
plt.title("Median JobSatPoints_6 by Experience Level")
plt.xlabel("Experience Level (Years)")
plt.ylabel("Median Job Satisfaction (JobSatPoints_6)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# --- Cerrar conexión ---
conn.close()
```
